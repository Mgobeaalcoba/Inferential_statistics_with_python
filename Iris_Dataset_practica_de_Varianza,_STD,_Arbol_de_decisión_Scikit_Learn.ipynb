```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

image.png

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
iris = pd.read_csv(url, names = names)
iris.head()
```

### Voy a armar un arbol de decisión con Scikit Learn que me permita entender si a traves del largo y ancho del sepalo y del petalo de las flores yo puedo deducir su clasificación y en que grado.

```python
# bibliotecas de machine learning
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn import tree
```

```python
iris.columns
```

```python
# guardar en variable X los atributos predictores (todas las etiquetas excepto "Sobreviviente")
X = iris.drop('class', axis=1) # dropeamos la unica columna que es dependiente para señalar por default a las independientes. axis=1 significa que es columna

# guardar en y la etiqueta a predecir ("Sobreviviente")
y = iris['class']
```

```python
X.head()
```

```python
y.head()
```

```python
# Creamos un objeto arbol
arbol = DecisionTreeClassifier(max_depth=3, random_state=42) # max_depth me marca los niveles de profundidad qu
```

```python
# entrenamos a la máquina
arbol.fit(X, y)
```

```python
# Predecimos sobre nuestro set
pred_y = arbol.predict(X)

# Comaparamos con las etiquetas reales
print('Precision: ', accuracy_score(pred_y, y))
```

```python
# creamos una matriz de confusión
confusion_matrix(y, pred_y)
```

```python
# creamos un gráfico para la matriz de confusión normalizado con los valores calculados sobre las columnas
plot_confusion_matrix = ConfusionMatrixDisplay.from_estimator(estimator= arbol,
                                                              X=X,
                                                              y=y,
                                                              cmap=plt.cm.Blues,
                                                              values_format='.2f',
                                                              normalize='pred')
plot_confusion_matrix
```

### Conclusiones:

1. Es practicamente infalible nuestro arbol de decisión para clasificar flores en función del largo y el ancho de sus petalos.
2. Solo 4 casos sobre 150 fueron mal clasificados por el arbol.

```python
# mostramos un árbol gráficamente
plt.figure(figsize= (10,8))
tree.plot_tree(decision_tree=arbol,
               filled=True,
               feature_names=X.columns)
plt.show()
```

```python
# graficamos las importancias en un gráfico de barras
# creamos las variables x (importancias) e y (columnas)
importancias = arbol.feature_importances_
columnas = X.columns

# creamos el gráfico
sns.barplot(x=columnas, y=importancias)
plt.title("Importancia de cada atributo")
plt.show()
```

#### ¿Como puedo saber yo si la desviación estandar es alta, media o baja? ¿Contra que lo puedo comparar? Claramente debo hacerlo contra la media. Viendo los dos valores calculo cuanto representa la desviación estandar contra la media, Si supera el 20% estamos en presencia de una muestra con un desvio estandar fuerte. Aunque este parametro va a depender mucho de lo que estemos analizando.

```python
iris['sepal-length'].mean()
```

```python
iris['sepal-length'].std(ddof=0) / iris['sepal-length'].mean()
```

En este caso el STD es el 14% del promedio. Por lo que siguiendo mi criterio podríamos decir que es tiene una desviación estandar media.

```python
# Creacion de una muestra aleatoria
muestra = iris.sample(frac=0.5)
muestra.head()
```

#### Veamos ahora que peso que valor asume el promedio, varianza y std en nuestra muestra:

```python
print(f"Promedio {muestra['sepal-length'].mean()}")
print(f"Varianza {muestra['sepal-length'].var(ddof=1)}")
print(f"Desviación Estandar {muestra['sepal-length'].std(ddof=1)}")
print(f"Relacion Desviación Estandar / Promedio {muestra['sepal-length'].std(ddof=1)/muestra['sepal-length'].mean()}")
```
