#### 2) Muestreo sistemático

```python
def systematic_sampling(data, step):
  indexes = np.arange(0, len(data), step=step)
  return econdata.iloc[indexes]

systematic_sample = systematic_sampling(econdata, 3)
systematic_sample # Va de 3 en 3 para seleccionar la muestrar.
```

#### Corregimos el "Cuauhtémoc" que está mal en la fuente

```python
def fix_words(column):
  for i in range(len(column)):
    if column[i] == "Cuauhtémoc":
      column[i] = "Cuautémoc"
  return column.copy()

econdata["delegacion"] = fix_words(econdata["delegacion"])
econdata["delegacion"].value_counts()
```

```python
econdata["estratificado"] = econdata["delegacion"] + "," + econdata["tipo"]
(econdata["estratificado"].value_counts()/len(econdata)*100).sort_values(ascending=False)
```

```python
def stratified_sample(data, category, size):
  strat_values = data[category].unique()
  strat_props = (data[category].value_counts() / len(data))
  strat_sample = pd.DataFrame(columns = data.columns)

  for value in strat_values:
    strat_sample = pd.concat([strat_sample, data[(data[category] == value)].sample(n = round(strat_props[value]*size)) ])

  return strat_sample
```

```python
df_estrat = stratified_sample(econdata, 'estratificado', 40)
df_estrat
```

```python
len(df_estrat)
```

```python
dict_pandas_series_muestreo = df_estrat["estratificado"].value_counts() / len(df_estrat)
dict_pandas_series_muestreo.plot(kind="bar")
```

#### Como se puede observar, la muestra conserva la proporción original de la población en cuanto al campo estratificado.
